```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.image as mpimg
```

```python

def draw_graph(G, with_labels=False, font_size=30, connection_style=0.1, savefig=False, file_format='pdf'):
      
    pos = nx.circular_layout(G, center=[0, 0], scale=1)
    rotate90 = np.array([[0, 1],
                         [-1, 0]])

    pos = {k: np.array([1, 1]) * (v @ rotate90) for k, v in pos.items()}
    colours = {k: sns.color_palette()[i] for i, k in enumerate(pos.keys())}
    print(colours)

    edge_weights = {k: np.exp(1.5*np.abs(v)) for k, v in nx.get_edge_attributes(G,'weight').items()}
    edge_styles = {k: '--' if v < 0 else '-' for k, v in nx.get_edge_attributes(G,'weight').items()}

    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    #fig = plt.figure(1)
    nx.draw(G, 
            ax=ax,
            with_labels=with_labels, 
            pos=pos, 
            node_size=7000, 
            node_color=list(colours.values()),
            font_size=font_size,
            font_family='Times New Roman',
            connectionstyle=f"arc3,rad={connection_style}",
            arrowstyle="-|>, head_length=1.8, head_width=0.6",
            width=list(edge_weights.values()),
            style=list(edge_styles.values()))


    ax.set_xlim(-1.3, 1.3)
    ax.set_ylim(-1, 1.4)
    plt.tight_layout()
    if savefig:
        plt.savefig(f'./plots/{savefig}.{file_format}', format=file_format, dpi=600)
    plt.show()
```

```python
graphs = {
    'common_cause': {
        'adjacency': np.array([[0, 1, 1],
                               [0, 0, 0],
                               [0, 0, 0]]),
        'labels': ['A', 'B', 'C'],
        'font_size': 30,
        'connection_style': 0,
        'file_name': 'common_cause',
        'file_format': 'png'
    },
    'collider': {
        'adjacency': np.array([[0, 0, 1],
                               [0, 0, 1],
                               [0, 0, 0]]),
        'labels': ['A', 'B', 'C'],
        'font_size': 30,
        'connection_style': 0,
        'file_name': 'collider',
        'file_format': 'png'
    },
    'standard_chain': {
        'adjacency': np.array([[0, 1, 0],
                               [0, 0, 1],
                               [0, 0, 0]]),
        'labels': ['A', 'B', 'C'],
        'font_size': 30,
        'connection_style': 0,
        'file_name': 'standard_chain',
        'file_format': 'png'
    },
    'confounded_chain': {
        'adjacency': np.array([[0, 1, 1],
                               [0, 0, 1],
                               [0, 0, 0]]),
        'labels': ['A', 'B', 'C'],
        'font_size': 30,
        'connection_style': 0,
        'file_name': 'confounded_chain',
        'file_format': 'png'
    },
    'damped_chain': {
        'adjacency': np.array([[0, 1, -1],
                               [0, 0, 1],
                               [0, 0, 0]]),
        'labels': ['A', 'B', 'C'],
        'font_size': 30,
        'connection_style': 0,
        'file_name': 'damped_chain',
        'file_format': 'png'
    },
    'damped_feedback': {
        'adjacency': np.array([[0, -1, -0.5],
                               [1, 0, -0.5],
                               [0, 0, 0]]),
        'labels': ['A', 'B', 'C'],
        'font_size': 30,
        'connection_style': 0.1,
        'file_name': 'damped_feedback',
        'file_format': 'png'
    },
    'damped_feedback_interA': {
        'adjacency': np.array([[0, -1, -0.5],
                               [0, 0, -0.5],
                               [0, 0, 0]]),
        'labels': ['A', 'B', 'C'],
        'font_size': 30,
        'connection_style': 0,
        'file_name': 'damped_feedback_interA',
        'file_format': 'png'
    },
    'damped_feedback_interB': {
        'adjacency': np.array([[0, 0, -0.5],
                               [1, 0, -0.5],
                               [0, 0, 0]]),
        'labels': ['A', 'B', 'C'],
        'font_size': 30,
        'connection_style': 0,
        'file_name': 'damped_feedback_interB',
        'file_format': 'png'
    },
    'damped_feedback_interC': {
        'adjacency': np.array([[0, -1, 0],
                               [1, 0, 0],
                               [0, 0, 0]]),
        'labels': ['A', 'B', 'C'],
        'font_size': 30,
        'connection_style': 0.1,
        'file_name': 'damped_feedback_interC',
        'file_format': 'png'
    },
    'damped_feedback_label': {
        'adjacency': np.array([[0, -1, -0.5],
                               [1, 0, -0.5],
                               [0, 0, 0]]),
        'labels': ['Police\nAction', 'Crime\nRate', 'Population\nHappiness'],
        'font_size': 18,
        'connection_style': 0.1,
        'file_name': 'damped_feedback_label',
        'file_format': 'png'
    },
}

for graph_name, graph_data in graphs.items():
    G = nx.from_numpy_array(graph_data['adjacency'], create_using=nx.DiGraph)

    if 'file_format' in graph_data.keys():
        file_format = graph_data['file_format']
    else:
        file_format = 'pdf'

    if 'labels' in graph_data.keys():
        mapping = {k:graph_data['labels'][i] for i, k in enumerate(G.nodes)}
        G = nx.relabel_nodes(G, mapping=mapping)
        with_labels = True
        font_size = graph_data['font_size']
    else:
        with_labels = False
        font_size = 0

    if 'file_name' in graph_data.keys():
        savefig = graph_data['file_name']
    else:
        savefig = False
    
    draw_graph(G, with_labels=True, font_size=font_size, connection_style=graph_data['connection_style'], savefig=savefig, file_format=file_format)

    img = mpimg.imread(f'./plots/{savefig}.png')
```
